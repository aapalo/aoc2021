```python
def monad(te, serial):
    vals = {}
    s = list(str(serial))
    for k in "xyzw":
        vals[k] = 0
    for i in te:
        i = i.split()
        if len(i) == 2:            
            if i[0] == "inp":
                vals[i[1]] = int(s.pop(0))
        else:
            if i[2] in "wxyz":
                v = int(vals[i[2]])
            else:
                v = int(i[2])
            if i[0] == "add":
                vals[i[1]] += v
            elif i[0] == "mul":
                vals[i[1]] *= v
            elif i[0] == "div":
                vals[i[1]] = int(vals[i[1]] / v)
            elif i[0] == "mod":
                vals[i[1]] = vals[i[1]] % v
            elif i[0] == "eql":
                vals[i[1]] = 1 if vals[i[1]] == v else 0
    #for i in (vals.keys()):
    #    print(i, vals[i])
    #print(vals.keys(), vals.values())
    return vals["z"]
```
